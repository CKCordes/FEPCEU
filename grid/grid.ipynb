# Gridmaker 4000

```python
from gridmaker import Grid
```

```python
grid = Grid((54.58, 6.87), (57.835613519, 13.4137049476), 100)
grid.show_grid()
```

```python
import json

points = grid.get_coords()

weather = {
    'endpoint_string': 'weather/point',
    'labels': ['time', 'airTemperature', 'windSpeed']
}
solar = {
    'endpoint_string': 'solar/point',
    'labels': ['time', 'uvIndex']
}
endpoints = {
    'weather': weather,
    'solar': solar
}


data = {
    'n_points': len(points),
    'time-period': ['01-01-2024', '31-12-2024'],
    'order': ['lat', 'lon'],
    'endpoints': endpoints,
    'points': points
}

with open('coordinates.json', 'w') as json_file:
    json.dump(data, json_file, indent=4)
```
